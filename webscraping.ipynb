```python
import requests
from bs4 import BeautifulSoup
import os
import zipfile

# Define the base URL
base_url = 'https://law.justia.com/codes/new-jersey/2022/title-15a/'

# Send an HTTP GET request and parse the HTML
reqs = requests.get(base_url)
soup = BeautifulSoup(reqs.text, 'html.parser')

# Create a directory to store the scraped content
output_directory = 'scraped_content'
os.makedirs(output_directory, exist_ok=True)

# Create a list to store the URLs
urls = []

# Extract and store all the URLs from the webpage
for link in soup.find_all('a', href=True):
    href = link.get('href')
    absolute_url = urljoin(base_url, href)
    urls.append(absolute_url)

# Loop through each URL and scrape data
for idx, url in enumerate(urls, start=1):
    try:
        # Send an HTTP GET request and parse the HTML of the current URL
        req = requests.get(url)
        page_soup = BeautifulSoup(req.text, 'html.parser')

        # Extract the part of the URL to create a filename
        url_parts = url.split('/')
        filename = "_".join(url_parts[-4:])

        # Save the HTML content to a file with the constructed filename
        file_name = f'{filename}.html'
        file_path = os.path.join(output_directory, file_name)

        with open(file_path, 'w', encoding='utf-8') as file:
            file.write(page_soup.prettify())

        print(f"Scraped and saved: {url} to {file_path}")
    except Exception as e:
        print(f"An error occurred while scraping {url}: {e}")

# Create a zip file containing all the HTML files
with zipfile.ZipFile('scraped_content.zip', 'w') as zf:
    for idx in range(1, len(urls) + 1):
        url_parts = urls[idx - 1].split('/')
        filename = "_".join(url_parts[-4:])
        file_name = f'{filename}.html'
        file_path = os.path.join(output_directory, file_name)
        zf.write(file_path, file_name)

print("All HTML files have been saved and zipped.")
```

```python
import zipfile

# Specify the path to the zip file
zip_file_path = 'scraped_content.zip'

# Initialize a count
file_count = 0

# Open the zip file and count its members (files)
with zipfile.ZipFile(zip_file_path, 'r') as zf:
    file_count = len(zf.namelist())

print(f"Number of files in {zip_file_path}: {file_count}")
```

```python
import os
import shutil

# Define the path to the zip file
zip_file_path = 'scraped_content.zip'

# Define the folder where you want to extract the contents
extracted_folder = 'extracted_content'

# Create the extracted folder if it doesn't exist
os.makedirs(extracted_folder, exist_ok=True)

# Extract the contents of the zip file
with zipfile.ZipFile(zip_file_path, 'r') as zf:
    zf.extractall(extracted_folder)

# Loop through the extracted files and rename HTML files to .txt
for root, dirs, files in os.walk(extracted_folder):
    for file in files:
        if file.endswith('.html'):
            old_path = os.path.join(root, file)
            new_path = os.path.join(root, file.replace('.html', '.txt'))
            os.rename(old_path, new_path)

print("HTML files have been renamed to .txt")

# Now, you have all the HTML files renamed to .txt in the 'extracted_content' folder.
```

```python
import zipfile
import os

# Define the folder with renamed .txt files
extracted_folder = 'extracted_content'

# Define the name of the output zip file
output_zip_file = 'renamed_content.zip'

# Create a new zip file and add the .txt files
with zipfile.ZipFile(output_zip_file, 'w', zipfile.ZIP_DEFLATED) as zf:
    for root, dirs, files in os.walk(extracted_folder):
        for file in files:
            if file.endswith('.txt'):
                file_path = os.path.join(root, file)
                arcname = os.path.relpath(file_path, extracted_folder)
                zf.write(file_path, arcname)

print("Renamed .txt files have been saved to a new zip file:", output_zip_file)
```
